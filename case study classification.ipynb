### MISSING VALUES

```python
#check for null values
df.isna().sum()
```

```python
df[df.isnull().any(axis =1)]
```

```python
# Missing values for each length can be filled with the median of each class.
median = df.groupby('Classification')['SL','SW','PL'].median()
median
```

```python
median.loc['Iris-setosa','SL']
```

```python
for c in ['Iris-setosa', 'Iris-versicolor', 'Iris-virginica']:
    for f in ['SL', 'SW', 'PL']:
        df.loc[df['Classification'] == c,f] = df[df['Classification'] == c][f].fillna(median.loc[c,f])
```

```python
df.isna().sum()
```

```python
df.describe()
```

```python
df.Classification.value_counts()
```

```python
classes = {'Iris-setosa':1,'Iris-versicolor':2, 'Iris-virginica':3 }
df['Classification'] = df['Classification'].map(classes)
```

```python
sns.heatmap(df.corr(),annot =True)
```

### OUTLIER HANDLING

Outlier is present in SW.Since SW is not contributing much to classification(also due to overlapped distribution among 3 classes) we will drop SW.

### X,y SPLIT

```python
# for X we will drop the SW also as it is not contributiong much to the classification
X = df.drop(['Classification','SW'], axis =1)
y = df['Classification']
```

### TRAIN TEST SPLIT

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split( X, y, test_size=0.2, random_state=1, stratify = y)
```

### Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report
lr = LogisticRegression()
lr.fit(X_train,y_train)
ypred = lr.predict(X_test)
print(classification_report(y_test,ypred))
```

### SVM
